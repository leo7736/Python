```python
from sklearn.datasets import load_iris
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
import numpy as np
```

```python
iris=load_iris()
```

Logistic regression

```python
l_score=cross_val_score(LogisticRegression(),iris.data,iris.target)
l_score
```

```python
np.average(l_score)
```

Decision Tree

```python
d_score=cross_val_score(DecisionTreeClassifier(),iris.data,iris.target)
d_score
```

```python
np.average(d_score)
```

SVM

```python
s_score=cross_val_score(SVC(),iris.data,iris.target)
s_score
```

```python
np.average(s_score)
```

Random Forest

```python
r_score=cross_val_score(RandomForestClassifier(n_estimators=35),iris.data,iris.target)
r_score
```

```python
np.average(r_score)
```
