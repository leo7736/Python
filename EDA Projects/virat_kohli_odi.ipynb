## **Getting Info**

 ## 1.finding against which country virat have highest batting avg?

```python
 kohli_odi.groupby('Opp').agg({'runs':'sum','Opp':'count'})
```

```python
dict1= {}
for i in kohli_odi.index:
  runs = kohli_odi.loc[i].runs
  country= kohli_odi.loc[i].Opp
  if country not in dict1:
    dict1[country]=[runs,1]
  else:
    dict1[country]= [dict1[country][0]+runs,dict1[country][1]+1]
```

```python
dict1
```

```python
maxx = 0
team= ''
for c,v in dict1.items():
  avg = v[0]/v[1]
  if avg>maxx:
    maxx = avg
    team =c
maxx,team
```

## 3.avg. SR of virat kohli whenever the score is between  90- 100 (100 excluded)?

```python
 avg_list=[]
 for i in kohli_odi.index:
   row = kohli_odi.loc[i]
   runs= row['runs']
   if runs >= 90 and runs<100:
     avg_list.append(float(row['sr']))
avg_sr= sum(avg_list)/len(avg_list)
avg_sr
```

### 4.Range where virat scored max six and four

```python
runs = kohli_odi['runs']
group_bin=[i //10 for i in runs]
kohli_odi['group_size']= group_bin
df1 = kohli_odi.groupby('group_size').sum().reset_index()
df1.head(20)
```

```python
max_six =0
for i in df1.index:
  row= (df1.iloc[i])
  six_run=row['sixes']
  if max_six<six_run:
    max_six = six_run
    max_four = row['fours']
    group_int = row['group_size']

max_six,max_four,group_int
```
