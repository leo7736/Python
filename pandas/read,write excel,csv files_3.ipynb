```python
def convert_people_cell (cell):
    if cell == 'n.a.':
        return 'sam walton'
    return cell

def convert_price_cell (cell):
    if cell == 'n.a.':
        return 50
    return cell

def convert_people_cell (cell):
    if cell == 'n.a.':
        return 'sam walton'
    return cell
def convert_revenue_cell (cell):
    if cell == -1:
        return 0
    return cell



df=pd.read_excel('stock_data.xlsx','Sheet1',converters= {
        'people': convert_people_cell,
        'price': convert_price_cell,
    'revenue':convert_revenue_cell
    })

df
```

```python
#write to excel
```

```python
df.to_excel('new.xlsx',sheet_name='stocks',index=False,startrow=1,startcol=3)
```

```python
df_stocks = pd.DataFrame({
    'tickers': ['GOOGL', 'WMT', 'MSFT'],
    'price': [845, 65, 64 ],
    'pe': [30.37, 14.26, 30.97],
    'eps': [27.82, 4.61, 2.12]
})

df_weather =  pd.DataFrame({
    'day': ['1/1/2017','1/2/2017','1/3/2017'],
    'temperature': [32,35,28],
    'event': ['Rain', 'Sunny', 'Snow']
})


df_weather
```

```python
with pd.ExcelWriter('stocks_weather.xlsx') as writer:
    df_stocks.to_excel(writer, sheet_name="stocks",index=False)
    df_weather.to_excel(writer, sheet_name="weather",index=False)
```
