```python
import pandas as pd
import numpy as np
df = pd.read_csv("weather_data.csv")
df
```

```python
new_df=df.replace(-99999,np.NaN)
new_df
```

```python
new_df=df.replace([-99999,-88888],np.NaN)
new_df
```

```python
new_df=df.replace({
    'temperature':-99999,
    'windspeed':-99999,
    'event':'0'
},np.NaN)
new_df
```

```python
#mapping values for replace
```

```python
new_df=df.replace({-99999:np.NaN,'0':'sunny'})
new_df
```

```python
df = pd.read_csv("weather_data.csv")
df
```

```python
new_df=df.replace({
    'temperature':'[A-Za-z]',
    'windspeed':'[A-Za-z]'},' ',
    regex=True)
new_df
```

```python
new_df
```

```python
new_df=new_df.replace(['Snow','Rain','Sunny'],[1,2,3])
new_df
```
